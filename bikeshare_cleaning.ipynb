# Get the bikeshare data ready for analysis

```python
import pandas as pd
import pickle
```

#### import 2016-Q3 bike data (add more data sets soon)

```python
bikes1 = pd.read_csv('2016-Q3-Trips-History-Data-1.csv')
bikes2 = pd.read_csv('2016-Q3-Trips-History-Data-2.csv')
bikes = pd.concat([bikes1, bikes2], ignore_index=True)
```

Create DatetimeIndex from `Start date`

```python
bikes['Start date'] = pd.to_datetime(bikes['Start date'], format='%m/%d/%Y %H:%M')
```

```python
bikes = bikes.set_index('Start date')
```

save this DataFrame with pickle

```python
pickle.dump( bikes, open( "bikeshare_rides.p", "wb" ) )
```
